Herramientas útiles en Databricks:

```
%python
# Lanzar código en Python:
lista = ['pelota','cubo','pala']
print("La variable lista: ",lista,"es del tipo:", type(lista))
```

```
%r
# Lanzar código en R
numero <- list('pelota','cubo','pala')
numero
print(paste0(numero))
```
